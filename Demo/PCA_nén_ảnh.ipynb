### IMAGE COMPRESSION USING PCA


```python
# Importing the required libraries:
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
```

```python
# Reading the Open Genus Logo as input
img = mpimg.imread('opengenus_logo.png')
```

```python
# Printing the shape of the image:
print(img.shape) 
#Output: 156 rows each containing 196 pixels which have 3 channels(RGB)

# Showing the image
plt.imshow(img)
```

```python
# Resize the image so that it is in the format required for PCA input 
# As 196*3 = 588, resize the image to (156,582)
img_r = np.reshape(img, (156, 582)) 
print(img_r.shape) 
```

```python
# Running PCA with 32 Principal Components
pca = PCA(32).fit(img_r) 
img_transformed = pca.transform(img_r) 
print(img_transformed.shape) 
print(np.sum(pca.explained_variance_ratio_) )
# Output: With these 32 components we are able to express 98.7% of the variance
```

```python
# Inverse transforming the PCA output and reshaping 
temp = pca.inverse_transform(img_transformed) 
print(temp.shape)
temp = np.reshape(temp, (156, 194 ,3)) 
print(temp.shape) 
```

```python
# Visualization using imshow
plt.imshow(temp)
```
